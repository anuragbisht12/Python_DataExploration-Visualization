```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#list the available styles and themes
print(plt.style.available)
```

```python
#change the style
plt.style.use('seaborn-pastel')
```

```python
tips = sns.load_dataset("tips")  # Load data
print(tips)
```

```python
#scatterplot
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
# Scatterplot with Seaborn
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, scatter_kws={'edgecolors': 'w'}, fit_reg=True)
plt.show()
```

```python
#line plots
sns.lineplot(x='size', y='total_bill', legend='full', data=tips , hue='smoker')
```

```python
#histograms
sns.distplot(tips['total_bill'], bins=10, kde=True)
plt.axvline(tips['total_bill'].mean(), color="k", linestyle="--");
```

```python
sns.kdeplot(tips['total_bill'])
plt.axvline(tips['total_bill'].mean(), color="k", linestyle="--",animated=True);
plt.text(tips['total_bill'].mean(),0,np.round(tips['total_bill'].mean(),4),rotation=270)

plt.axvline(tips['total_bill'].median(), color="k", linestyle=":",animated=True);
plt.text(tips['total_bill'].median(),0,np.round(tips['total_bill'].median(),4),rotation=270)
```

```python
#density plots
#with sns.axes_style('white'):
sns.jointplot("total_bill", "tip", tips, kind='kde');
```

```python
#violenplots
sns.violinplot("size", "tip", hue="smoker",data=tips,palette=["lightblue", "lightgreen"],split=True, inner="quartile");
```

```python
#count plots
sns.countplot(tips['size'])
```

```python
#finding corelation using heatmaps
sns.heatmap(tips.corr(), annot=True)
```

```python
#pair plots of all measures
sns.pairplot(tips)
```
